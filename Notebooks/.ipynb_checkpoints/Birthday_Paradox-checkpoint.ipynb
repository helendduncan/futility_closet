# The Birthday Paradox


The birthday paradox is that in a group of just 23 people, there is over a 50 % chance that two people will share a birthday.

There are some subtlties here that I am not touching on, but the gist of this notebook is to test the paradox by brute force methods. Nothing sophisticated here!

To break it down into the bare bones of the problem: I am stating that when picking 23 random numbers between 1 and 365, there is a 50% chance that I will pick the same number at least once.


So what do I need to test this problem:


1. [Random Number](#RandNumb)


### RandNumb:

First lets get a random number generator: Numpy will do but we should mention that it's not really a true random number generator.

What we want this to do is randomly pick a number between 1 and 365 from a flat distribution. This means that if I picked say 10,000 random numbers using randint and generated a histogram of the results, I would expect the probablitiy of getting any given number to be about the same as the probability of getting any other number

```python
def check_uniformity(ndata):
    """
    A method to generate a list of numbers between 1 and 365
    4 lists are produced with length ndata, 10ndata, 100ndata and 1000ndata
    The probability of any particular value is plotted
    """
    
    fig, ax = plt.subplots(4,1, figsize = (18,22))
    for i in range(4):
        npoints = ndata*(10**i)
        x=np.random.randint(1,366,npoints)
        hist = ax[i].hist(x, density=True, bins = 365)
        ax[i].set_xlabel('Day')
        ax[i].set_ylabel('Probability')
        ax[i].title.set_text(f'Distribution of {npoints} points')
        ax[i].axhline((ndata/365)/ndata, color='r', linestyle='-')
        ax[i].set_xlim(1, 365)
    plt.show()
```

```python
check_uniformity(1000)
```

```python
check_uniformity(10000)
```

```python
check_uniformity(100000)
```

```python
check_uniformity(1000000)
```
